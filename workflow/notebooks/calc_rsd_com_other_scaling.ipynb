```python
import pandas as pd
```

```python
base_demands=pd.read_csv(snakemake.input[0]).set_index("DateUTC")
rsd_com_other_dem=(pd.read_csv(snakemake.input.rsd_com_other_dem)
                   .replace({"UK":"GB"})
                   .set_index("ISO2"))
```

```python
tot_base=base_demands.sum()/1E6

scaling=tot_base.add(rsd_com_other_dem.squeeze()).div(tot_base)
```

```python
scaled_demands=base_demands.mul(scaling,axis=1)

scaled_demands.reset_index().to_csv(snakemake.output[0],index=False)
```
