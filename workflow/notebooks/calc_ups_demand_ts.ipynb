```python
import pandas as pd
import numpy as np
```

```python
year=int(snakemake.wildcards.year)
countries=snakemake.params.countries

h2=(pd.read_csv(snakemake.input.h2_annual)
    .replace({"UK":"GB"})
    .fillna(0))

syn=(pd.read_csv(snakemake.input.syn_annual)
    .replace({"UK":"GB"})
    .fillna(0))

d=(h2.merge(syn,left_on="ISO2",right_on="ISO2",how="outer")
   .fillna(0)
   .set_index("ISO2")
   .sum(axis=1))
```

```python
dates=pd.date_range(start=pd.Timestamp(year,1,1,0),end=pd.Timestamp(year,12,31,23),freq="h")

dem=pd.DataFrame(np.ones((dates.shape[0],len(countries)))*(1/dates.shape[0]),index=dates,columns=countries)
dem=dem.mul(d)*1E6
```

```python
dem.to_csv(snakemake.output[0])
```
